Elbow method


```python
inertia = []
for i in range(1,11):
  print(i)
  km = KMeans(i)
  km.fit(scaled_df)
  print(km.inertia_)
  inertia.append(km.inertia_)
```

```python
inertia
```

```python
no_of_clusters = range(1,11)
plt.plot(no_of_clusters, inertia, marker = "o")
plt.xlabel("no of clusters")
plt.ylabel("inertia")
plt.title("Elbow curve")
plt.show()
```
